Choosing pixels for omparisong between observed spectra and 'best-match' model spectra:

- **Centrally Reversed Spectra**: [8,50,150], [8,61,171], [13,25,154]

```python
file_name = "spectroscopic_data/iris_l2_20140528_112458_3820111493_raster_t000_r00008.fits" # 

raster8 = fits.open(file_name, memmap=False, do_not_scale_image=False) # raster files are largers so memmap is a good idea
print(raster8[1].data.shape)


plt.figure(figsize=[20,10])
plt.imshow(raster8[1].data[..., 66], vmin=-200, vmax=800)
x1 = [190] #[190,180]
y1 = [50] #[55,30]
x2 = [150,171] #[156,154]
y2 = [50,61] #[34,25]
x3 = [141,140]
y3 = [61,46]
plt.plot(x1,y1,'x', color='r', label='single peaked')
plt.plot(x2,y2,'x', color='b', label='centrally reversed')
plt.plot(x3,y3,'x', color='y', label='complex')
plt.legend()
plt.grid()
plt.show()
```
